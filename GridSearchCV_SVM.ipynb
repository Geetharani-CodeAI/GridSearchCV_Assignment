```python
import numpy as np
import matplotlib.pyplot
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','sigmoid'],
              'C':[10,100,1000,2000,3000],'gamma':['scale','auto']}
grid = GridSearchCV(SVR(), param_grid, refit = True,verbose = 3, n_jobs = -1)
grid.fit(independent,dependent)
```

```python
print(grid.best_params_)
re = grid.cv_results_
print(re)
```

```python
print("The best R_score value Parameters:{}".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
age = float(input("Age:"))
bmi = float(input("BMI:"))
children = float(input("Children:"))
sex_male = int(input("sex_male 0 or 1:"))
smoker_yes = int(input("smoker_yes 0 or 1"))
Future_Prediction = grid.predict([[age,bmi,children,sex_male,smoker_yes]])
print("Future_Prediction={}".format(Future_Prediction))
```
